```python
meeting_title = "Sample Meeting Title"
logo_str = "https://via.placeholder.com/100x35?text=Logo"  # placeholder logo URL
meeting_info_html = "<p>Meeting info content here.</p>"
html_linked_summary = "<p>Summary content here.</p>"
financial_snapshot_html = "<p>Financial snapshot here.</p>"
html_client_industry_sources = "<p>Industry sources here.</p>"
html_client_holdings_sources = "<p>Client holdings here.</p>"
html_macro_news_sources = "<p>Macro news here.</p>"

full_html = f"""
<html>
<head>
    <meta charset='UTF-8'>
    <title>{meeting_title}</title>
    <style>
        body,p, ul, ol, li, div {{
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            font-size: 10px;
            margin: 30px auto;
            color: #333;
            line-height: 1.6;
            max-width: 800px;
        }}
        /* other CSS unchanged */
    </style>
</head>
<body>
    <div class="header-container">
        <img src="{logo_str}" alt="Logo" class="logo" />
        <h1>{meeting_title}</h1>
    </div>

    <h2>Meeting Information</h2>
    {meeting_info_html}

    <div class="feedback-section">
        {html_linked_summary}
        <div class="feedback-buttons">
            <button onclick="handleFeedback(this, 'more')">I want to see more of this</button>
            <button onclick="handleFeedback(this, 'less')">I do not want to see this</button>
        </div>
        <div class="feedback-message" style="display:none; margin-top:5px; color: green;">
            Got it! We'll remember it next time.
        </div>
    </div>

    <h2>Macro-Economic Snapshot</h2>
    <div class="feedback-section box-content">
        {financial_snapshot_html}
        <div class="feedback-buttons">
            <button onclick="handleFeedback(this, 'more')">I want to see more of this</button>
            <button onclick="handleFeedback(this, 'less')">I do not want to see this</button>
        </div>
        <div class="feedback-message" style="display:none; margin-top:5px; color: green;">
            Got it! We'll remember it next time.
        </div>
    </div>

    <h2>Resources</h2>
    <div class="feedback-section box-content">
        {html_client_industry_sources}
        {html_client_holdings_sources}
        {html_macro_news_sources}
        <div class="feedback-buttons">
            <button onclick="handleFeedback(this, 'more')">I want to see more of this</button>
            <button onclick="handleFeedback(this, 'less')">I do not want to see this</button>
        </div>
        <div class="feedback-message" style="display:none; margin-top:5px; color: green;">
            Got it! We'll remember it next time.
        </div>
    </div>

<script>
function handleFeedback(button, type) {{
    const feedbackSection = button.closest('.feedback-section');
    const buttons = feedbackSection.querySelector('.feedback-buttons');
    const message = feedbackSection.querySelector('.feedback-message');

    buttons.style.display = 'none';
    message.style.display = 'block';

    setTimeout(() => {{
        message.style.display = 'none';
        buttons.style.display = 'flex';  // restore buttons after message hides
    }}, 5000);
}}
</script>

</body>
</html>
"""

# Save and open in browser
with open("preview.html", "w", encoding="utf-8") as f:
    f.write(full_html)
```
