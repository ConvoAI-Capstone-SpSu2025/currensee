## Hasbro

```python
df1 = pd.read_sql("""SELECT * FROM meeting_data 
where host_email ='jane.moneypenny@bankwell.com' and invitee_emails = 'jessica.palmer@hasbro.com'
order by meeting_timestamp desc""", con=engine)
df1.head(30)
```

```python
with engine.begin() as connection:  # ensures COMMIT
    connection.execute(text("""
        UPDATE meeting_data 
        SET meeting_subject = 'Hasbro - Discuss Portfolio Diversification and '
        WHERE host_email = 'jane.moneypenny@bankwell.com'
          AND invitee_emails = 'jessica.palmer@hasbro.com'
          AND TRIM(meeting_timestamp) = '2025-07-10 14:30:00' 
    """))
```

```python
df = pd.read_sql("""SELECT *
        FROM email_data
        WHERE to_emails in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com') 
        and from_email in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com') 
        order by email_timestamp desc""", con=engine)
df.head(10)
```

```python
with engine.begin() as connection:  # ensures COMMIT
    connection.execute(text("""
        UPDATE email_data
        SET email_body = 'Hi jane,  Yes, this needs further thought. I can do 1-5pm tomorrow, or 9am on the 13th. Let me know which works for you. Regards, Jessica Palmer, VP Hasbro  Email: jessica.palmer@hasbro.com  Phone: 909.878.8329x31984'
        WHERE to_emails in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com')
          AND from_email in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com')
          AND TRIM(email_timestamp) = '2025-07-09 16:10:00' 
    """))
```

```python
with engine.begin() as connection:  # ensures COMMIT
    connection.execute(text("""
        UPDATE email_data
        SET email_timestamp = '2025-07-05 13:25:00'
        WHERE to_emails = 'jessica.palmer@hasbro.com'
          AND from_email = 'jane.moneypenny@bankwell.com'
          AND TRIM(email_timestamp) = '2024-01-20 14:40:00'
    """))
```

```python
df = pd.read_sql("""SELECT email_timestamp
        , to_names	
        , to_emails
        , from_name
        , from_email
        , email_subject
        , email_body
        FROM email_data
        WHERE to_emails in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com') 
        and from_email in ('jane.moneypenny@bankwell.com', 'jessica.palmer@hasbro.com') 
        and email_timestamp < '2025-04-30'
        order by email_timestamp desc""", con=engine)
df.head(20)
```
