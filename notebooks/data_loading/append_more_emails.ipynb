```python
crm_outlook_engine = create_pg_engine('crm_outlook')
```

```python
query= ''' select * from email_data where (to_emails = 'jessica.palmer@hasbro.com' or from_email = 'jessica.palmer@hasbro.com')'''
email_df = pd.read_sql(query, crm_outlook_engine)
```

```python
create = text("""
CREATE TABLE email_data2 AS
SELECT * FROM email_data;
""")
with crm_outlook_engine.connect() as connection:
    connection.execute(create)
    connection.commit()
```

```python
query= ''' select * from email_data2 '''
email_df2 = pd.read_sql(query, crm_outlook_engine)
```

# Append one extra meeting for jane and jessica

```python
create = text("""
CREATE TABLE meeting_data2 AS
SELECT * FROM meeting_data;
""")

with crm_outlook_engine.connect() as connection:
    connection.execute(create)
    connection.commit()
```

```python
query= ''' select * from meeting_data2 where invitee_emails = 'jessica.palmer@hasbro.com' '''
meeting_df = pd.read_sql(query, crm_outlook_engine)
meeting_df.head(30)
```

```python
update= '''UPDATE meeting_data2
SET meeting_subject = 'Hasbro - Discuss Portfolio Diversification and Bond Strategy'
WHERE meeting_timestamp = '2024-01-30 10:00:00'
  AND invitee_emails = 'jessica.palmer@hasbro.com'
  '''

with crm_outlook_engine.connect() as connection:
    connection.execute(text(update))
    connection.commit()
```
